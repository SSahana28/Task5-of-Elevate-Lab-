```python
import pandas as pd

# Load data
train = pd.read_csv("Downloads/titanic/train.csv")
gender_submission = pd.read_csv("Downloads/titanic/gender_submission.csv")

# Info and summary
print(train.info())
print(train.describe(include="all"))

# Value counts for categorical features
print(train['Sex'].value_counts())
print(train['Embarked'].value_counts())
print(train['Pclass'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot
sns.pairplot(train[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.show()

# Heatmap of correlations

numeric_cols = train.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()

```

```python
# Survival by Sex
sns.barplot(x='Sex', y='Survived', data=train)
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Probability")
plt.show()

# Survival by Pclass
sns.barplot(x='Pclass', y='Survived', data=train)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Probability")
plt.show()

# Survival by Embarked
sns.barplot(x='Embarked', y='Survived', data=train)
plt.title("Survival Rate by Embarkation Point")
plt.ylabel("Survival Probability")
plt.show()
```

```python
train.dropna(subset=['Age', 'Fare'], inplace=True)
sns.set_style('whitegrid')

#Histograms
train['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

train['Fare'].hist(bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

#Boxplots
sns.boxplot(x='Survived', y='Age', data=train)
plt.title("Age vs Survived")
plt.show()

sns.boxplot(x='Survived', y='Fare', data=train)
plt.title("Fare vs Survived")
plt.show()

#Scatterplot
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train)
plt.title("Age vs Fare by Survival")
plt.legend(loc='upper right')  # Optional: change location if overlapping
plt.show()
```
